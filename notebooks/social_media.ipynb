```python
import pandas as pd
import numpy as np

# Load the two survey datasets
df_qatariinfluencer = pd.read_csv("Final Data.csv") 
df_instagraminfluencer = pd.read_csv("Instagram Influencers effect digital marketing_2022-2023.csv")

# 1.Dataset 1: Copy
df_qatariinfluencer_clean = df_qatariinfluencer.drop_duplicates().copy()

# 2.Dataset 2: Copy
df_instagraminfluencer_clean = df_instagraminfluencer.copy()

# MAPPING DICTIONARIES
GENDER_MAP = {1: 'Male', 2: 'Female', 3: 'Other'}
AGE_MAP_1 = {1: '18-24', 2: '25-34', 3: '35-44', 4: '45-54', 5: '55-64', 6: '65+'}
INCOME_MAP = {1: '<â‚¬18k', 2: 'â‚¬18k-â‚¬30k', 3: 'â‚¬30k-â‚¬42k', 4: 'â‚¬42k-â‚¬54k', 5: '>â‚¬54k'}
FOLLOWERS_MAP = {1: '1-100', 2: '101-500', 3: '501-1k', 4: '1k-5k', 5: '>5k'}
INFLUENCERS_MAP = {1: '0-1', 2: '2-5', 3: '6-10', 4: '11-20', 5: '>20'}
FREQUENCY_MAP = {1: 'Rarely', 2: 'Weekly', 3: 'Daily'}
def map_age_group(age):
    if pd.isna(age) or age <= 17: return 'Unknown'
    if age in range(18, 25): return '18-24'
    if age in range(25, 35): return '25-34'
    if age in range(35, 45): return '35-44'
    if age >= 45: return '45+'
    return 'Unknown'


# PROCESS DATASET 1

# 1. Map Demographic Columns to Labels
df_qatariinfluencer_clean['Gender_Label'] = df_qatariinfluencer_clean['gender'].map(GENDER_MAP).fillna('Unknown')
df_qatariinfluencer_clean['Age_Group_Label'] = df_qatariinfluencer_clean['age'].map(AGE_MAP_1).fillna('Unknown')
df_qatariinfluencer_clean['Income_Label'] = df_qatariinfluencer_clean['income'].map(INCOME_MAP).fillna('Unknown')
df_qatariinfluencer_clean['Nationality_Label'] = df_qatariinfluencer_clean['nationality'].map({1: 'Spain', 2: 'Italy', 3: 'Portugal', 4: 'Other'}).fillna('Unknown')

# 2. Define Likert Columns and ensure they are numeric
DEMO_COLS_ORIGINAL_1 = ['gender', 'age', 'nationality', 'income']
DEMO_COLS_LABEL_1 = ['Gender_Label', 'Age_Group_Label', 'Nationality_Label', 'Income_Label']
LIKERT_COLS_1 = [col for col in df_qatariinfluencer_clean.columns 
                 if col not in DEMO_COLS_ORIGINAL_1 and col not in DEMO_COLS_LABEL_1]

# 3. Explicitly convert to numeric to prevent issues
for col in LIKERT_COLS_1:
    df_qatariinfluencer_clean[col] = pd.to_numeric(df_qatariinfluencer_clean[col], errors='coerce')

# 4. Aggregate Data by Demographics
DEMO_COLS_TO_GROUP_1 = ['Gender_Label', 'Age_Group_Label', 'Income_Label'] 

df_agg1 = df_qatariinfluencer_clean.groupby(DEMO_COLS_TO_GROUP_1)[LIKERT_COLS_1].mean(numeric_only=False).reset_index()

# 5. Rename columns to be descriptive
df_agg1.columns = DEMO_COLS_TO_GROUP_1 + [f"Avg_{col}" for col in LIKERT_COLS_1]


# PROCESS DATASET 2

# CRITICAL FIX: Convert Age and Likert columns to numeric first
df_instagraminfluencer_clean['Age'] = pd.to_numeric(df_instagraminfluencer_clean['Age'], errors='coerce')

# 1. Map Demographic/Engagement Columns to Labels
df_instagraminfluencer_clean['Age_Group_Label'] = df_instagraminfluencer_clean['Age'].apply(map_age_group)
df_instagraminfluencer_clean['Followers_Group'] = df_instagraminfluencer_clean['Instagram_Followers_numbers'].map(FOLLOWERS_MAP).fillna('Unknown')
df_instagraminfluencer_clean['Influencers_Group'] = df_instagraminfluencer_clean['Instagram_Influencers_numbers'].map(INFLUENCERS_MAP).fillna('Unknown')
df_instagraminfluencer_clean['Frequency_Group'] = df_instagraminfluencer_clean['Instagram_frequency_use'].map(FREQUENCY_MAP).fillna('Unknown')

# 2. Identify Likert Columns for Averaging
LIKERT_COLS_2 = [
    'Comparison_with_Influencers', 'Likelihood_buy_product_Infleucer_recommendation', 'Envy', 
    'AdvertisementScept', 'LifeSatisfactio', 'MaternalismSuccess', 'MaternalismCentrality', 
    'MaternalismHapiness', 'SocialComparison'
] 

# 3. Explicitly convert Likert columns to numeric
for col in LIKERT_COLS_2:
    df_instagraminfluencer_clean[col] = pd.to_numeric(df_instagraminfluencer_clean[col], errors='coerce')

# 4. Aggregate Data by Engagement/Demographics
DEMO_COLS_TO_GROUP_2 = ['Age_Group_Label', 'Followers_Group', 'Frequency_Group']

df_agg2 = df_instagraminfluencer_clean.groupby(DEMO_COLS_TO_GROUP_2)[LIKERT_COLS_2].mean(numeric_only=False).reset_index()

# 5. Rename columns to be descriptive
df_agg2.columns = DEMO_COLS_TO_GROUP_2 + [f"Avg_{col}" for col in LIKERT_COLS_2]

# --- Final Step: Save aggregated dataframes to CSV for review ---
df_agg1.to_csv('aggregated_survey1_features.csv', index=False)
df_agg2.to_csv('aggregated_survey2_features.csv', index=False)


print(f"File 1: aggregated_survey1_features.csv")
print(f"File 2: aggregated_survey2_features.csv")
print("\n--- Aggregated Survey 1 (df_agg1) Head ---")
print(df_agg1.head().to_markdown(index=False))
print("\n--- Aggregated Survey 2 (df_agg2) Head ---")
print(df_agg2.head().to_markdown(index=False))
```
