```python
def normalize_accords(accords):
    # If the value is already a list
    if isinstance(accords, list):
        return accords
    # Handle the string representation of an empty list
    elif accords == '[]' or pd.isna(accords):
        return []
    # If it's a string representation of a list ('[citrus, floral]')
    elif isinstance(accords, str) and accords.startswith('['):
        try:
            return [n.strip().lower() for n in ast.literal_eval(accords)]
        except:
            return [] # Return empty list if parsing fails
    # If it's a simple comma-separated string ('citrus, floral')
    elif isinstance(accords, str):
        return [n.strip().lower() for n in accords.split(',')]
    return []
```

```python
df_perfume_clean['Main Accords'] = df_perfume_clean['Main Accords'].apply(normalize_accords)
df_perfume_clean
```

```python
# Combine Top/Middle/Base Notes into Main Accords

def combine_notes(row):
    notes = []
    for col in ['Top Notes', 'Middle Notes', 'Base Notes']:
     if pd.notna(row[col]):
      notes.extend([n.strip().lower() for n in str(row[col]).split(',')])
    return notes

df_parfumo_clean['Main Accords'] = df_parfumo_clean.apply(combine_notes, axis=1)
```

```python
# The master list of all columns
columns_keep = ['perfume_clean', 'brand_clean', 'type_clean', 'target_audience',
                'longevity', 'Release Year','Rating Value', 'Rating Count', 'Main Accords']

# Use .filter() to select only the columns that actually exist in each DataFrame.

df_perfume_final = df_perfume_clean.filter(columns_keep)
df_fragrance_final = df_fragrance_clean.filter(columns_keep)
df_parfumo_final = df_parfumo_clean.filter(columns_keep)

# Check the resulting column sets to verify
print("Perfume Columns:", df_perfume_final.columns.tolist())
print("Fragrance Columns:", df_fragrance_final.columns.tolist())
print("Parfumo Columns:", df_parfumo_final.columns.tolist())
```

```python
# Concatenate datasets

df_combined = pd.concat([df_perfume_final, df_fragrance_final, df_parfumo_final],axis=0, ignore_index=True)

# Optional: remove exact duplicates

df_combined = df_combined.drop_duplicates(subset=['perfume_clean', 'brand_clean'])

# Reset index

df_combined = df_combined.reset_index(drop=True)
```

```python
# Flatten the List for Final ML Feature Engineering
# Convert the list back to a string for easier NLP processing later.
df_combined['Main Accords String'] = df_combined['Main Accords'].apply(lambda x: ' '.join(x))
```
